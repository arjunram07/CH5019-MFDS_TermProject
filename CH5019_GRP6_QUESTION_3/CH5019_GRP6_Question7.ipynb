```python
# converting ICMRTestingData to df

testing_df = pd.read_csv("/Users/arsh/MFDS-Project/Term-project-2020/Dataset_Question3/ICMRTestingDetails.csv")
```

```python
testing_df
```

```python
cases_till_req_date = testing_df.ix[24, 'TotalPositiveCases']
```

```python
cases_till_req_date
```

```python
tests_per_day_per_lab = 100
tests_per_ten_days_per_lab = tests_per_day_per_lab*10
```

```python
rate = 0.1
```

```python
rate_of_positive_tests = 0.046
```

```python
number_of_labs = ((cases_till_req_date*((1 + rate)**10) - cases_till_req_date)*(1/rate_of_positive_tests))/tests_per_ten_days_per_lab
```

```python
number_of_labs = round(number_of_labs)
```

```python
number_of_labs
```

```python
number_of_additional_labs = number_of_labs - (testing_df.ix[24, 'TotalSamplesTested'] - testing_df.ix[23, 'TotalSamplesTested'])/tests_per_day_per_lab
```

```python
number_of_additional_labs = round(number_of_additional_labs)
```

```python
number_of_additional_labs
```

```python
labs_reqd = pd.DataFrame({'Cases till 10/04/20' : [cases_till_req_date], 'Cases till 20/04/20 at given rate' : [round(cases_till_req_date*((1 + rate)**10))], 'Number of Active Labs on 10/04/2020' : [round((testing_df.ix[24, 'TotalSamplesTested'] - testing_df.ix[23, 'TotalSamplesTested'])/tests_per_day_per_lab)],'Number of Additional Labs Required on 20/04/20' : [number_of_additional_labs]})
```
